# Which sector receives the largest share of fixed capital investments?

```python
data=%sql SELECT AVG(AGR), AVG(MIN),AVG(MAN),AVG(ENE),AVG(TRCOM),AVG(TOU),AVG(HOU),AVG(EDU),AVG(HEA),AVG(OTSER) FROM capital;
data
```

```python
data=np.array(data).reshape(10,)
cat=np.array(['AGR','MIN','MAN','ENE','TRCOM','TOU','HOU','EDU','HEA','OTSER']).reshape(10,)
```

```python
fig,ax= plt.subplots(figsize=(15, 6))
plt.bar(cat, data)
plt.xlabel('Sectors')
plt.ylabel('Shares')
plt.title('Bar Chart of Averages')
plt.show()
```

# Showing The Sectors Having Big Share 

```python
YILLAR = %sql SELECT YILLAR FROM capital;
YILLAR=pd.DataFrame(YILLAR).astype(int)
MAN = %sql SELECT MAN FROM capital;
MAN=pd.DataFrame(MAN).astype(float)
```

```python
def make_scat(y):
    fig,ax= plt.subplots(figsize=(15, 6))
    plt.scatter(YILLAR, y)
    plt.axhline(y.mean().values, color='red', linestyle='--', label='Average')
    plt.xticks(np.arange(1963,2016))
    plt.xticks(rotation='vertical')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.ylabel('Share')
    plt.legend()
    plt.show()
```

## The Scatter Graph of Capital Investment on Manufacturing

```python
make_scat(MAN)
```

# 1990-2001 Crisis Years

```python
c=%sql SELECT YILLAR,MAN,AGR,TRCOM,HOU from capital where YILLAR between 1990 and 2001;
c=pd.DataFrame(c,columns=['Years','Man','Agr','TrCom','Hou'])
```

```python
c
```

```python
fig,ax= plt.subplots(figsize=(15, 6))
plt.plot(c['Years'],c['Man'],color='red', linestyle='--',label='Manufacturing')
plt.plot(c['Years'],c['Agr'],color='blue', linestyle='--',label='Agriculture')
plt.plot(c['Years'],c['Hou'],color='green', linestyle='--',label='Housing')
plt.plot(c['Years'],c['TrCom'],color='brown', linestyle='--',label='TransportationCommunication')
plt.legend()
plt.show()
```

```python
d=%sql SELECT YILLAR,EDU,HEA from capital where YILLAR between 1990 and 2001;
d=pd.DataFrame(d,columns=['Years','Edu','Hea'])
```

```python
d
```

```python
fig,ax= plt.subplots(figsize=(15, 6))
plt.plot(d['Years'],d['Edu'],color='red', linestyle='--',label='Education')
plt.plot(d['Years'],d['Hea'],color='blue', linestyle='--',label='Health')
plt.legend()
plt.show()
```
